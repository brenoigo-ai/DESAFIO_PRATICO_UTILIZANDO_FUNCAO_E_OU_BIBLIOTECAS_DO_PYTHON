1. Jogo de Adivinhação. Utilizando a bibliteca **random** crie um pequeno jogo de adivinhação de números inteiros que deve ir de 1 a 100. As regras do jogo devem ser:



*   Todos os palpites do usuário devem ser armazenados
*   O programa deve informar que o usuário errou o palpite e mostrar uma dica se o número a ser descoberto é maior ou menor do que o número do palpite
*   Quando o usuário acertar o palpite o programa deve mostrar: todos os números digitados para o palpite, a quantidade de tentativa




```python
import random

def jogo_adivinhacao():
  print('PIAUÍ INSTITUTO DE TECNOLOGIA - PIT')
  print('Bacharelado em Inteligência Artificial')
  print()
  print('Jogo de Adivinhação')
  print()
  print('Regras: Tente adivinhar o número entre 1 e 100.')
  print('Ao fazer seu palpite o sistema irá informar se acertou ou se errou.')
  print('No segundo caso, o jogo revelará se o número a ser descoberto é maior ou menor do que o seu palpite.')
  print('Tente acertar no menor número de tentativas')
  print()

  nick = input('Digite seu nickname: ')
  numero_secreto = random.randint(1, 100)
  tentativas = []

  while True:
    palpite = int(input('Digite seu palpite entre 1 e 100: '))

    if palpite < 1 or palpite > 100:
      print('Valor inválido! Digite um número entre 1 e 100')
      continue

    tentativas.append(palpite)

    if palpite > numero_secreto:
      print('Você errou o palpite! O número a ser descoberto é MENOR.')
      print()
    elif palpite < numero_secreto:
      print('Você errou o palpite! O número a ser descoberto é MAIOR.')
      print()
    elif palpite == numero_secreto:
      print()
      print(f'Parabéns, {nick}! Você acertou o número secreto!')
      print(f'Você acertou em {len(tentativas)} tentativas!')
      print(f'Seus palpites foram: {tentativas}')
      break
    else:
      print('Algo deu errado!')

jogo_adivinhacao()
```

2. O professor de matemática Alan Turing pediu sua ajuda  para criar um pequeno sistema para calcular as áreas das  figuras planas. Ele pediu que o usuário possa escolher qual figura quer calcular. Para cada cálculo de uma figura faça uma função específica. Utilize a figura abaixo como apoio matemático e das figuras que seu programa deve conter.

image.png

```python
import math

def area_triangulo (b,h):
  return (b*h)/2

def area_quadrado (l):
  return l**2

def area_losango (d1,d2):
  return (d1*d2)/2

def area_retangulo (b,h):
  return b*h

def area_trapezio (b1,b2,h):
  return ((b1+b2)*h)/2

def area_circulo (r):
  return math.pi*r**2

print('Bacharelado em Inteligência Artificial - PIT')
print('ÁREA DE FIGURAS PLANAS')
print()

while True:
  print()
  print('Escolha qual a figura plana que deseja calcular a área:')
  area = int(input('1-TRIÂNGULO \n2-QUADRADO \n3-LOSANGO \n4-RETÂNGULO \n5-TRAPÉZIO \n6-CÍRCULO \n0-SAIR \n'))

  if area == 0:
    print('Até a próxima!')
    print('Bacharelado em Inteligência Artificial - PIT')
    break

  if area < 0 or area > 6:
    print('Digite uma opção válida!')
    continue

  print()
  print('Necessário que as medidas estejam na mesma unidade!')

  match area:
    case 1:
      print('ÁREA DO TRIÂNGULO')
      print()
      b = float(input('Digite o valor da base: '))
      h = float(input('Digite o valor da altura: '))
      print(f'({b})*({h})/2 = {area_triangulo(b,h)}')
    case 2:
      print('ÁREA DO QUADRADO')
      print()
      l = float(input('Digite o valor do lado: '))
      print(f'{l}^2 = {area_quadrado(l)}')
    case 3:
      print('ÁREA DO LOSANGO')
      print()
      d1 = float(input('Digite o valor da diagonal maior: '))
      d2 = float(input('Digite o valor da diagonal menor: '))
      print(f'({d1})*({d2})/2 = {area_losango(d1,d2)}')
    case 4:
      print('ÁREA DO RETÂNGULO')
      print()
      b = float(input('Digite o valor da base: '))
      h = float(input('Digite o valor da altura: '))
      print(f'{b}*{h} = {area_retangulo(b,h)}')
    case 5:
      print('ÁREA DO TRAPÉZIO')
      print()
      b1 = float(input('Digite o valor da base maior: '))
      b2 = float(input('Digite o valor da base menor: '))
      h = float(input('Digite o valor da altura: '))
      print(f'({b1}+{b2})*{h}/2 = {area_trapezio(b1,b2,h)}')
    case 6:
      print('ÁREA DO CÍRCULO')
      print()
      r = float(input('Digite o valor do raio: '))
      print(f'pi*{r}^2 = {area_circulo(r):.2f}')
    case _:
      print('Opção inválida!')
```

3. Utilizando função e o módulo **math** crie um programa para calcular as possíveis raízes de uma equação do 2º grau. Sabendo que: ax² + bx + c = 0. Determine e mostre o número de raízes reais distintas que a equação possui. Regra: O número de raízes reais depende do discriminante (delta), *delta* = b² - 4ac:

image.png

*   *delta* > 0: Duas raízes reais distintas;
*   *delta* = 0: Uma raiz real (ou duas raízes reais iguais);
*   *delta* < 0: Nenhuma raiz real (duas raízes complexas).

```python
import math

def delta(a, b, c):
  return b**2 - 4*a*c

def raiz(a, b, valor_delta):
  x1 = (-b + math.sqrt(valor_delta))/(2*a)
  x2 = (-b - math.sqrt(valor_delta))/(2*a)
  return x1, x2

print('PIAUÍ INSTITUTO DE TECNOLOGIA - PIT')
print('Bacharelado em Inteligência Artificial')
print()
print('RAÍZES DE UMA EQUAÇÃO DE 2º GRAU')

while True:
  a = float(input('Digite o valor de a: '))
  b = float(input('Digite o valor de b: '))
  c = float(input('Digite o valor de c: '))
  print()

  valor_delta = delta(a, b, c)

  if valor_delta > 0:
    print('DUAS RAÍZES REAIS DISTINTAS')
    print(f'Raízes: {raiz(a, b, valor_delta)}')
  elif valor_delta == 0:
    print('UMA RAIZ REAL (OU DUAS RAÍZES REAIS IGUAIS)')
    print(f'Raiz: {raiz(a, b, valor_delta)}')
  elif valor_delta < 0:
    print('NENHUMA RAIZ REAL (DUAS RAÍZES COMPLEXAS)')
  else:
    print('Algo deu errado!')

  print()
  continuar = input('Deseja calcular a raiz de outra equação de segundo grau? (S/N): ').upper()

  while True:
    if continuar == 'S' or continuar == 'N':
      break
    else:
      print('Opção inválida!')
      continuar = input('Deseja calcular a raiz de outra equação de segundo grau? (S/N): ').upper()

  if continuar == 'N':
    print()
    print('Até a próxima!')
    print('PIAUÍ INSTITUTO DE TECNOLOGIA - PIT')
    print('Bacharelado em Inteligência Artificial')
    break
```

4. Vimos na nossa aula sobre Expressões Regulares (RegEx), o módulo re em Python. Sabendo disso pesquise mais sobre os recursos desse módulo e faça um programa que:

*   Importe re
*   Crie uma função extrair_telefones(texto)
*   Use re.findall() para encontrar números de telefone no formato: (XX) XXXXX-XXXX
*   Caso não encontre, informe que não encontrou

Exemplo:
Qual seu número de telefone

Meu número é (86) 99999-1234 e (86) 99999-9876

Saída: [‘(86) 99999-12334’, ‘(86) 99999-9876’]

Retorne todos os números encontrados

```python
import re

def extrair_telefones(texto):
  codigo = r'\(\d{2}\) \d{5}-\d{4}'
  telefones_encontrados = re.findall(codigo, texto)

  if len(telefones_encontrados) == 0:
    return 'Nenhum número de telefone encontrado.'
  else:
    return telefones_encontrados

print('PIAUÍ INSTITUTO DE TECNOLOGIA - PIT')
print('Bacharelado em Inteligência Artificial')
print()
print('Encontrando telefones usando RegEx')

texto = input('Olá! Tudo bem? Fale um pouco sobre você e me diga qual seu número de telefone? Por favor, informe nesse formato: (xx) xxxxx-xxxx')
print()

telefones_encontrados = extrair_telefones(texto)
print(f'Telefones encontrados: {telefones_encontrados}')

```

5. Essa questão você deve criar seu próprio módulo, ou seja, a questão deverá ter dois arquivos, 1 para o módulo e outro para o código principal. Imagine que você trabalhe na parte de Cyber segurança de uma empresa e precisa testar se as senhas criadas pelos os usuários estão no padrão da empresa. O seu módulo deverá ter uma função que verifique as senhas pelos critérios:

* Pelo menos 8 caracteres

* Contém pelo menos uma letra maiúscula

* Contém pelo menos um número

* Caso os critérios não sejam seguidos informar ao usuário

Dica: utilize no seu módulo a biblioteca **re**

```python
%%writefile seguranca.py
import re

def verificar_senha(senha):
  erros = []
  if len(senha) < 8:
    erros.append('Senha deve ter pelo menos 8 caracteres.')
  if not re.search('[A-Z]', senha):
    erros.append('Senha deve conter pelo menos uma letra maiúscula.')
  if not re.search('[0-9]', senha):
    erros.append('Senha deve conter pelo menos um número.')

  if len(erros) == 0:
    return True, 'Senha válida'
  else:
    return False, f'Sua senha não atinge o padrão de segurança exigido pela empresa! Verifique os erros: {erros}'
```

```python
import seguranca

print('PIAUÍ INSTITUTO DE TECNOLOGIA - PIT')
print('Bacharelado em Inteligência Artificial')
print()
print('VERIFICAÇÃO DE SENHA')
print()

while True:
  senha_digitada = input('Colaborador, digite sua senha (mínimo de 8 caracteres, necessário pelo menos um número e uma letra maiúscula): ')

  status, mensagem = seguranca.verificar_senha(senha_digitada)

  if status == True:
    print()
    print(f'{mensagem}')
    break
  elif status == False:
    print()
    print(f'{mensagem}')
    print('Por favor, tente novamente')
  else:
    print()
    print('Algo deu errado!')
    print('Por favor, tente novamente')
    break
```
